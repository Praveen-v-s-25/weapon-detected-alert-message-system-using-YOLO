```python
import yaml

# Specify the path to your data.yaml file
yaml_path = data_yaml

# Specify the new paths for train and validation datasets
new_train_path = train_images
new_val_path = val_images

try:
    # Load the YAML file
    with open(yaml_path, 'r') as file:
        data = yaml.safe_load(file)

    # Update the paths
    data['train'] = new_train_path
    data['val'] = new_val_path

    # Save the updated YAML file
    with open(yaml_path, 'w') as file:
        yaml.safe_dump(data, file)

    print("Updated data.yaml successfully!")
    print(data)
except FileNotFoundError:
    print(f"File not found: {yaml_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```
